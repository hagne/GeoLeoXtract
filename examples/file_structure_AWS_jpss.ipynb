# AWS and the s3fs library

This is an introduction into how to use the s3fs library for acessing the AWS system ... In case you don't want to use the nesdis_aws package or simply want to understand the key library in it.

## connect to file system

```python
import s3fs
```

```python
aws = s3fs.S3FileSystem(anon=True)
```

# explore file system

```python
base_folder = pl.Path(f'noaa-jpss')
```

## satellites

```python
satellites_available = aws.glob(base_folder.joinpath('*').as_posix())
satellites_available
```

## sensors

```python
satellite = 'NOAA20'
```

```python
sensors_available = aws.glob(base_folder.joinpath(satellite).joinpath('*').as_posix())
sensors_available
```

## products

```python
sensor = 'VIIRS'
```

```python
products_available = aws.glob(base_folder.joinpath(satellite).joinpath(sensor).joinpath('*').as_posix())
products_available
```

```python
len(products_available)
```

## month

```python
year = '2022'
```

```python
month_available = aws.glob(base_folder.joinpath(satellite).joinpath(sensor).joinpath(product).joinpath(year).joinpath('*').as_posix())
month_available
```

## days

```python
month = '10'
```

```python
days_available = aws.glob(base_folder.joinpath(satellite).joinpath(sensor).joinpath(product).joinpath(year).joinpath(month).joinpath('*').as_posix())
days_available
```
